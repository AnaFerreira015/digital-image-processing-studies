## Versão adaptada para Google Colab

```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```python
def generate_image(seed, width, height, mean, std):
    """
    Generates a grayscale image with pixel values sampled from a normal distribution.

    Args:
        seed (int): Random seed for reproducibility (student's registration number).
        width (int): Width of the generated image.
        height (int): Height of the generated image.
        mean (float): Mean of the normal distribution.
        std (float): Standard deviation of the normal distribution.

    Returns:
        image (numpy.ndarray): The generated image.
    """
    # Define a semente para reprodutibilidade
    np.random.seed(seed)

    # Gera a imagem com distribuição normal e converte para o intervalo [0, 255]
    image = np.random.normal(loc=mean, scale=std, size=(height, width))
    image = np.clip(image, 0, 255).astype(np.uint8)

    return image
```

```python
registration_number = 2025100636
width = 512
height = 512
mean = 128
std = 30
output_path = 'second_image_generated.png'
```

```python
image = generate_image(registration_number, width, height, mean, std)
cv2_imshow(image)
```

```python
cv2.imwrite(output_path, image)
print(f"Imagem gerada e salva em {output_path}")
```
